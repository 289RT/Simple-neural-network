```python
# 生成数据集
# 生成带噪点的y=sinx函数数据
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成x值，范围从0到10π
x = np.linspace(0, 10 * np.pi, 10000)

# 计算正弦值
y = np.sin(x)

noise = np.random.normal(0, 0.08, x.shape)

y_noisy = y + noise
```

```python
# 绘制正弦函数
plt.plot(x, y_noisy, label='Noisy Sine Function')
plt.plot(x, y, label='Original Sine Function', linestyle='--')
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
data = np.column_stack((x, y_noisy))
np.savetxt('sine_with_noise.txt', data, header='x\ty_noisy', fmt='%0.5f', delimiter='\t')
```
